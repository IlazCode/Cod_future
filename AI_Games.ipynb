```python
left = 1
right = int(input())
while True:
    current = (left+right)//2
    is_right = input('Ваше число:{}?(да = 1, больше = 2, меньше = 3)'.format(current))
    if is_right.lower() == '1':
        print('Угадал!')
        break
    elif is_right=='2':
        left = current + 1
    else:
        right = current - 1
```

```python
def prognos(left, right, d =2):
  n = np.random.randint(left, right)
  print('Ваше число:{}'.format(int(n)))
  pase =0
  while True:
    pase +=1
    current = (left+right)//d
    #print(current, "New")
    if current==n:
      print('Угадал! Число = ', current)
      print('Шагов',pase)
      break
    elif current<n:
      print(current, "Меньше")
      left = current + 1
    else:
      print(current, "Больше")
      right = current - 1
```

```python
def prognos1(left, right, d =2):
  n = np.random.randint(left, right)
  print('Ваше число:{}'.format(int(n)))
  pase =0
  while True:
    pase +=1
    #current = math.ceil((left+right)/d)
    current = math.floor((left+right)/d)
    #print(current, "New")
    if current==n:
      print('Угадал! Число = ', current)
      print('Шагов',pase)
      break
    elif current<n:
      print(current, "Меньше")
      left = current + 1
    else:
      print(current, "Больше")
      right = current - 1
```
