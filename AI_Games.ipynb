"Угадай число"
Пользователь загадывает число от 1 до N. Программа запрашивает у пользователя N. Требуется отгадать загаданное число, задав как можно меньше вопросов. (Обычно подобные задачи решают с помощью цикла).

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
left = 1
right = int(input())
while True:
    current = (left+right)//2
    is_right = input('Ваше число:{}?(да = 1, больше = 2, меньше = 3)'.format(current))
    if is_right.lower() == '1':
        print('Угадал!')
        break
    elif is_right=='2':
        left = current + 1
    else:
        right = current - 1
```

```python
left=1
right=100
d=2
n = np.random.randint(left, right, size=1)
print('Ваше число:{}'.format(int(n)))
pase =0
past = []
while True:
  pase +=1
  current = np.random.randint(left, right)

  if current not in past:
    #print(past)
    past.append(current)

    if current==n:
      print('Угадал! Число = ', current)
      print('Шагов',pase)
      break
    elif current<n:
      print(current, "Меньше")
    else:
      print(current, "Больше")
```

```python
def prognos(left, right, d =2):
  n = np.random.randint(left, right)
  print('Ваше число:{}'.format(int(n)))
  pase =0
  while True:
    pase +=1
    current = (left+right)//d
    #print(current, "New")
    if current==n:
      print('Угадал! Число = ', current)
      print('Шагов',pase)
      break
    elif current<n:
      print(current, "Меньше")
      left = current + 1
    else:
      print(current, "Больше")
      right = current - 1
```
