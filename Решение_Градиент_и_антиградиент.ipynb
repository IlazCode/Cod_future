# Градиент

Градиент - это вектор своим модулем равный скорости роста функции, а направлением указывающий направление наиболее быстрого её роста. Антиградиент в свою очередь указывает направление наиболее быстрого убывания. Мы уже определяли экстремум функции для функции одной переменной, это было достаточно наглядно и просто, немного сложнее это было для функции двух пенременных, когда мы использовали метод наименьших квадратов. В случае трёхмерного пространства можно предстаить горный рельеф и градиеннт в каждой точки горного склона будет указывать на направление вверх и по величине будет равен крутизне подъёма, антиградиент наоборот будет описывать спуск в ущелье.

Чаще всего нас будет интересовать именно минимум функции, он используется в очень популярном алгоритме машинного обучения -градиентный спуск.

Для начала необходимо вспомнить, что производная $f'$ с которой мы знакомились может быть названа производной по направлению, так как она характеризует скорость изменения функции $f= (x)$ по направлению оси $OX$. 


В методе наименьших квадратов у нас возникала функция двух переменных $F(a,b)$ и для нахождения минимума нам пришлось поочерёдно фиксируя переменные вычислять производную сначала по одному, а потом по другому направлению. 

Представив функцию в трёхмерном пространстве $z = f(x,y)$ увидим, что для определения скорости изменения высоты придётся вычислить производные по направлениям $OX$ и $OY$ и решить систему уравнений для нахождения их минимума(максимума).

Так же возможно вычислить производную (при условии её существования) по направлению произвольного луча $l$ на плоскости в выбранной точке по формуле  $\frac{\partial z}{\partial l} = z_x'(x;y)cos \alpha+z_y'(x;y)cos\beta$ при этом можно сказать, что при

$\frac{\partial z}{\partial l}>0$ - функция стремится подняться к "вершине горы"


$\frac{\partial z}{\partial l}<0$  - функция стремится спуститься в "ущелье"

$\frac{\partial z}{\partial l}=0$ - функция постоянна и находится на "плато"

Рассмотрим градиент для трёхмерного пространства. Размерность вектора градиента определяется размерностью пространства. Поэтому в данном представляет собой сумму частных производных по имеющимся трём координатным осям. 

$$\triangledown f = \frac{\partial f}{\partial x}i+ \frac{\partial f}{\partial y}j+ \frac{\partial f}{\partial z}k$$

**Пример**

Найти производную функции $z=y^3-x^2y+2xy$ в точке $A(2;-1)$ по направлению координатных осей и вектора $\overline{a}=-3\overline{i}-4\overline{j}$ (или записав сокращённо $\overline{a}(-3;-4)$)

Найдём  производные по направлениям осей координат

$z_x'=(y^3-x^2y+2xy)_x'=0-2xy+2y $

$z_y'=(y^3-x^2y+2xy)_y'=3y^2-x^2+2x $

Определим значения производных в избранной точке $A(2;-1)$

$z_x'(A)=z_x'(2;-1)=4-2=2$

$z_y'(A)=z_y'(2;-1)=3-4+4=3$

Определим производную по направлению вектора  $\overline{a}(-3;-4)$), для этого найдём единичный вектор, координаты которого и являются направляющими косинусами данного направления

$\overline{a} =\frac{\overline{a}}{|\overline{a}|}=\frac{-3\overline{i}-4\overline{j}}{\sqrt{(-3)^2+(-4)^2}}=\frac{-3\overline{i}-4\overline{j}}{\sqrt{25}}=-\frac{{3}}{5}\overline{i}-\frac{{4}}{5}\overline{j}$

теперь вычислим непосредственно производную в заданной точке

$\frac{\partial z}{\partial l} = z_x'(x;y)cos \alpha+z_y'(x;y)cos\beta = -\frac{{3}}{5}*2-\frac{{4}}{5}*3= -\frac{{18}}{5} = -3.6$ 

Градиентом функции  в точке  называется направленный отрезок , отложенный от точки $A(x;y)$ , который показывает направление и скорость наискорейшего роста функции  в данной точке.

$\triangledown z(A)= z_x'(A)\overline{i}+z_y'(A)\overline{j}=2\overline{i}+3\overline{j}$

Производная по некоторому направлению  в точке  – это проекция градиента в данной точке на данное направление, при этом мы можем определить "наклон" функции в данной точке $arctg (-3.6)\approx -74.5^o$

# Задание

Найти производную функции $z=x-y$   в точке $A(0;0)$ по направлению координатных осей и вектора $\overline{a}(2;2)$

## Решение



Найдём  производные по направлениям осей координат

$z_x'=(x-y)_x'=1 $

$z_y'=(x-y)_y'=-1$

Определим значения производных в избранной точке $A(0;0)$

$z_x'(A)=z_x'(0;0)=1$

$z_y'(A)=z_y'(0;0)=-1$

Определим производную по направлению вектора  $\overline{a}(2;2)$), для этого найдём единичный вектор, координаты которого и являются направляющими косинусами данного направления

$\overline{a} =\frac{\overline{a}}{|\overline{a}|}=\frac{2\overline{i}+2\overline{j}}{\sqrt{(2)^2+(2)^2}}=\frac{2\overline{i}+2\overline{j}}{2\sqrt{2}}=\frac{{1}}{\sqrt{2}}\overline{i}+\frac{{1}}{\sqrt{2}}\overline{j}$

теперь вычислим непосредственно производную в заданной точке

$\frac{\partial z}{\partial l} = z_x'(x;y)cos \alpha+z_y'(x;y)cos\beta = \frac{{1}}{\sqrt{2}}*1+\frac{{1}}{\sqrt{2}}*(-1)= 0$ 

Градиент функции  в точке  $A(0;0)$ 

$\triangledown z(A)= z_x'(A)\overline{i}+z_y'(A)\overline{j}=\overline{i}-\overline{j}$


# Код для построения графиков

```python
#https://habr.com/ru/post/547424/
import numpy as np
import matplotlib.pyplot as plot


radius = 8                                  # working plane radius
global_epsilon = 0.000000001                # argument increment for derivative
centre = (global_epsilon, global_epsilon)   # centre of the working circle
arr_shape = 100                             # number of points processed / 360
step = radius / arr_shape                   # step between two points


def differentiable_function(x, y):
    #return np.sin(x+5) * np.exp((1 - np.cos(y)) ** 2) + np.cos(y*2-8) * np.exp((1 - np.sin(x)) ** 2) + (x - y) ** 2
    return -(np.sin(np.sqrt(x**2+y**2)))/(np.sqrt(x**2+y**2))
    
def rotate_vector(length, a):
    return length * np.cos(a), length * np.sin(a)


def derivative_x(epsilon, arg):
    return (differentiable_function(global_epsilon + epsilon, arg) -
            differentiable_function(epsilon, arg)) / global_epsilon


def derivative_y(epsilon, arg):
    return (differentiable_function(arg, epsilon + global_epsilon) -
            differentiable_function(arg, epsilon)) / global_epsilon


def calculate_flip_points():
    flip_points = np.array([0, 0])
    points = np.zeros((360, arr_shape), dtype=bool)
    cx, cy = centre

    for i in range(arr_shape):
        for alpha in range(360):
            x, y = rotate_vector(step, alpha)
            x = x * i + cx
            y = y * i + cy
            points[alpha][i] = derivative_x(x, y) + derivative_y(y, x) > 0
            if not points[alpha][i - 1] and points[alpha][i]:
                flip_points = np.vstack((flip_points, np.array([alpha, i - 1])))

    return flip_points


def pick_estimates(positions):
    vx, vy = rotate_vector(step, positions[1][0])
    cx, cy = centre
    best_x, best_y = cx + vx * positions[1][1], cy + vy * positions[1][1]

    for index in range(2, len(positions)):
        vx, vy = rotate_vector(step, positions[index][0])
        x, y = cx + vx * positions[index][1], cy + vy * positions[index][1]
        if differentiable_function(best_x, best_y) > differentiable_function(x, y):
            best_x = x
            best_y = y

    for index in range(360):
        vx, vy = rotate_vector(step, index)
        x, y = cx + vx * (arr_shape - 1), cy + vy * (arr_shape - 1)
        if differentiable_function(best_x, best_y) > differentiable_function(x, y):
            best_x = x
            best_y = y

    return best_x, best_y


def gradient_descent(best_estimates, is_x):
    derivative = derivative_x if is_x else derivative_y
    best_x, best_y = best_estimates
    descent_step = step
    value = derivative(best_y, best_x)

    while abs(value) > global_epsilon:
        descent_step *= 0.95
        best_y = best_y - descent_step \
            if derivative(best_y, best_x) > 0 else best_y + descent_step
        value = derivative(best_y, best_x)

    return best_y, best_x


def find_minimum():
    return gradient_descent(gradient_descent(pick_estimates(calculate_flip_points()), False), True)


def get_grid(grid_step):
    samples = np.arange(-radius, radius, grid_step)
    x, y = np.meshgrid(samples, samples)
    return x, y, differentiable_function(x, y)


def draw_chart(point, grid):
    point_x, point_y, point_z = point
    grid_x, grid_y, grid_z = grid
    plot.rcParams.update({
        'figure.figsize': (4, 4),
        'figure.dpi': 200,
        'xtick.labelsize': 4,
        'ytick.labelsize': 4
    })
    ax = plot.figure().add_subplot(111, projection='3d')
    ax.scatter(point_x, point_y, point_z, color='red')
    ax.plot_surface(grid_x, grid_y, grid_z, rstride=5, cstride=5, alpha=0.7)
    plot.show()


if __name__ == '__main__':
    min_x, min_y = find_minimum()
    minimum = (min_x, min_y, differentiable_function(min_x, min_y))
    draw_chart(minimum, get_grid(0.05))
```
