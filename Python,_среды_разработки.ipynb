# Установка программного обеспечения.

Интерпретатор языка можно бесплатно скачать интерпретатор Python по адресу https://python.org/ , на сайте присутствуют различные версии для Linux, MacOs и Windows 

Так же возможно установить сразу пакет Anaconda, находящийся по адресу https://www.anaconda.com/products/distribution данный продукт установит не только интерпретатор Python, но и даст возможность в единой оболочке загружать другие необходимые для работы с языкрм Python приложения.

image.png

Если вы желаете скачать только интерпретатор, то скачайте установщик Python 3 адаптивный для системы, установленной на вашем компьютере, он имеет расширение exe, и запустите его.

Вторым шагом лучше отключить ограничение длины MAX_PATH. (в системах Linux этих ограничений нет) Проигнорировав данный пункт, можно столкнуться с проблемой совместимости (код созданный на Linux может не запустится на Windows).

Проверить качество установки можно через командную строку открыв её в меню Пуск или запустив интерпретатор командной строки cmd.exe через диспетчер задач или поиск. Если после этого ввести команду Python то в результате командная строка выведет версию языка, которая установлена в данный момент системе 

image.png

Работать с интерпретатором конечно же можно используя IDLE (интегрированная среда разработки и обучения на языке Python) или Python Shell, но это очень ограниченные инструменты. Для упрощения работы с интерпретатором используют различные IDE (или интегрированная среда разработки) или редакторы кода в зависимости от предпочтений разработчика. Они гибко настраиваются и дают большое количество дополнительных сервисов, позволяющих сделать процесс работы максимально комфортным. К наиболее популярным решениям относятся Visual Studio Code , PyCharm , Spyder. Для исследований в единоборствах максимально подходящим может считаться Spyder , так как он изначально рассматривался как инструмент для Data Science. Также хорошо работает интеграция с IPython/Jupyter, что мы рассмотрим ниже. 

Оптимально для работы сразу установить пакет Anaconda дистрибутив, включающий языки программирования Python и R, набор популярных свободных библиотек и прикладных программ в том числе свободно распространяемых, объединённых проблематиками науки о данных и машинного обучения.
