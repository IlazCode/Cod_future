# Некоторые сведения из статистики и теории вероятности

Если мы попытаемся определить точность попадания в игрового персонажа броском снаряда (который был смоделирован ранее), то то после нескольких атак можно сделать заключение о точности (хорошо известного параметра в компьютерном спорте - Aim). Обращаясь к теории вероятности можно представить каждое из попаданий в качестве случайной величины. Действительно, мы не можем гарантировать попадание и всегда есть вероятность промаха большего или меньшего.
Для начала по определению классической вероятности можем определить вероятность попадания отношением количества попаданий в интересующую нас точку к общему количеству бросков.

$${\displaystyle \Pr(A)={\frac {n}{N}}}$$

Итак рассмотрим двух игроков.  Каждый пытается выполнив 4 броска попасть в персонажа (NPC). Мы рассмотрим отклонение от центра "мишени". Если игрок попал точно, то отклонений равно 0, если правее мишени на 5 единиц, то результат равен 5, а если левее, то -5.

|index|попытка\_1|попытка\_2|попытка\_3|попытка\_4|
|---|---|---|---|---|
|Player\_1|12|-6|7|-8|
|Player\_2|22|-16|17|-18|

«Снайперское» математическое ожидание равно , однако и у «интересной личности»:  – оно тоже нулевое!

Таким образом, возникает потребность количественно оценить, насколько далеко  рассеяны пули (значения случайной величины) относительно центра мишени (математического ожидания). Ну а рассеяние с латыни переводится не иначе, как дисперсия.

Теперь нам надо понять куда попадёт игрок при следующем броске. Не зная этого точно мы можем прогнозировать попадание и для этого используется математическое ожидание, которое означает среднее (взвешенное по вероятностям возможных значений) значение случайной величины. 

$${\displaystyle \mathbb {M} [x]=\sum \limits _{i=1}^{\infty }x_{i}\,p_{i}}$$

В нашем случае, это значит, что было сделано четыре попытки, а значит вероятность попадания в каждую из точек попадания  было равно одной четверти (0.25). Чтобы найти математическое ожидание (сделать прогноз) нам нужно сложить результаты разделённые на количество попыток. 

$${\displaystyle \mathbb {M} [Player_1]={\frac {12}{4}}-{\frac {6}{4}}+{\frac {7}{4}}-{\frac {8}{4}}}={\frac {12-6+7-8}{4}=1.25}$$

Как можно заметить математическое ожидание для подобных ситуаций равно средену арифметическому значению. И в результате прогноза мы получили ожидаемое отклонение при следующем броске от "центра мишени".

Второй игрок промахивается на большее значение, но давайте рассчитаем математическое ожидание для него и увидим, что оно не отличается от презультата для первого игрока.

$${\displaystyle \mathbb {M} [Player_2]={\frac {22}{4}}-{\frac {16}{4}}+{\frac {17}{4}}-{\frac {18}{4}}}={\frac {22-16+17-18}{4}=1.25}$$

Но результаты каждой отдельной попыкти у игроков отличаются тем что они на разное расстояние отстоят от центра. Поэтому кроме среднего значения рассматриваются также меры изменчивости.

Первый игрок более меткий и его результаты укладываются в отрезок $[-8;12]$ радиусом 20, а второй менее и его результаты описываются большим отрезком $[-18;22]$ радиусом 40. Таким образом первый игрок в два раза точнее второго.

Размах (range, разброс) выборки измеряет расстояние, в пределах которого изменяются оценки, разность между максимальным и минимальным значениями ряда.$$R = X_{(n)} - X_{(1)}$$ 


Для определения разброса значений случайной величины относительно её математического ожидания была введена дисперсия. Малая дисперсия означает, что значения сгруппированы близко друг к другу.
Большая дисперсия свидетельствует о сильном разбросе значений.


$${\displaystyle D[х]=\sum _{i=1}^{n}{p_{i}(x_{i}-\mathbb {M} [x])^{2}}}$$

Формула похожа на математическое ожидание, только мы вместо точки попадания, которая характеризовала удаление от центра мишени, мы используем удаление от среднего значения или мат. ожидания возведённое в квадрат. Если не произвести возведения в квадрат, то в результате мы получим 0, так как все точки равноудалены от среднего значения. 

$${\displaystyle Сум.отклонений [Player_1]={\frac {12-1.25}{4}}-{\frac {6-1.25}{4}}+{\frac {7-1.25}{4}}-{\frac {8-1.25}{4}}}={\frac {10.75-7.25+5.75-9.25}{4}=0}$$
Дополнительно вводится поправка (несмещённая оценка дисперсии, именно так будут считать методы в библиотеках Python) и делить нужно на $n-1$, что в нашем случае равно 3.
$${\displaystyle D [Player_1]={\frac {(12-1.25)^2}{3}}-{\frac {(6-1.25)^2}{3}}+{\frac {(7-1.25)^2}{3}}-{\frac {(8-1.25)^2}{3}}=95.58}$$

Но сама величина дисперсии, как отклонение от среднего не позволяет судить об этом отклонении в единицах его измерения, так как мы имеем квадрат этих отклонений. Поэтому вводят другой показатель рассеяния (дисперсии) случайной величины относительно её математического ожидания (средней величины)-  среднеквадратическое отклонение или стандартное отклонение.

Среднеквадратическое отклонение измеряется в единицах измерения самой случайной величины и используется при расчёте стандартной ошибки среднего арифметического, при построении доверительных интервалов, при статистической проверке гипотез, при измерении линейной взаимосвязи между случайными величинами.
 
$${\displaystyle \sigma ={\sqrt {D[X]}}={\sqrt {{\frac {1}{n-1}}\sum _{i=1}^{n}\left(x_{i}-{\bar {x}}\right)^{2}}}}$$.

Дисперсия и среднеквадратическое отклонение так же неустойчивы к выбросам, как и среднее арифметическое, поэтому иногда используют медиану или наиболее часто встречающееся значение (моду).

В случае необходимости можно нормировать данные, например разделить наш набор данных на стандартное отклонение и получить в результате вместо абсолютных значений количество стандартных отклонений ($σ$) в данной точке, тем самым можно достичь необходимого "улучшения" данных.
